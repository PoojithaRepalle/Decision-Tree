```python
import numpy as np
import pandas as pd
def importdata(path='ENJOYSPORT.csv'):
    data = pd.read_csv(path)
    return data

print(importdata())
```

```python
from sklearn.preprocessing import LabelEncoder

# Sample dataset structure
data = {
    'Sky': ['Sunny', 'Sunny', 'Rainy', 'Sunny'],
    'AirTemp': ['Warm', 'Warm', 'Cold', 'Warm'],
    'Humidity': ['Normal', 'High', 'High', 'High'],
    'Wind': ['Strong', 'Strong', 'Strong', 'Weak'],
    'Water': ['Warm', 'Warm', 'Warm', 'Cool'],
    'Forecast': ['Same', 'Same', 'Change', 'Same'],
    'EnjoySport': [1, 1, 0, 1]
}

# Convert to DataFrame
df = pd.DataFrame(data)

# Initialize LabelEncoder
label_encoder = LabelEncoder()

# Apply label encoding to all categorical columns
for column in df.columns[:-1]:  # Exclude the target column 'EnjoySport'
    df[column] = label_encoder.fit_transform(df[column])

# View the transformed dataset
print(df)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
import matplotlib.pyplot as plt

# Sample data
data = {
    'Sky': ['Sunny', 'Sunny', 'Rainy', 'Sunny'],
    'AirTemp': ['Warm', 'Warm', 'Cold', 'Warm'],
    'Humidity': ['Normal', 'High', 'High', 'High'],
    'Wind': ['Strong', 'Strong', 'Strong', 'Weak'],
    'Water': ['Warm', 'Warm', 'Warm', 'Cool'],
    'Forecast': ['Same', 'Same', 'Change', 'Same'],
    'EnjoySport': [1, 1, 0, 1]
}

# Convert to DataFrame
df = pd.DataFrame(data)

# Encode categorical features
label_encoder = LabelEncoder()
for column in df.columns[:-1]:
    df[column] = label_encoder.fit_transform(df[column])

# Separate features and target
X = df.drop('EnjoySport', axis=1)
y = df['EnjoySport']

# Create and train the Decision Tree Classifier
clf = DecisionTreeClassifier(criterion='entropy', random_state=0)
clf.fit(X, y)
```

```python
plt.figure(figsize=(12,8))
tree.plot_tree(clf, feature_names=X.columns, class_names=['No', 'Yes'], filled=True, rounded=True, fontsize=10)
plt.show()
```
